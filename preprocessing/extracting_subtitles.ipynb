```python
import webvtt
import os
import json
```

```python
# Function to parse .vtt subtitle files and extract relevant information
def parse_vtt_subtitles(vtt_file):
    # Create a list to hold the parsed subtitles
    subtitles = []
    # Read the .vtt file using webvtt
    for caption in webvtt.read(vtt_file):
        # Extract the start time, end time, and text from each caption
        start_time = caption.start
        end_time = caption.end
        text = caption.text.strip()
        # Append the extracted information to the subtitles list
        subtitles.append({
            'file': vtt_file.split("\\")[-1],
            'start': start_time,
            'end': end_time,
            'text': text
        })
    # Return the list of parsed subtitles
    return subtitles
```

```python
# Function to save normalized keypoints data to a JSON file
def save_to_json(data, file_path):
    with open(file_path, 'w') as f:
        json.dump(data, f)


folder_path = "<path_to_vtt_files>"
input_directory = "<path_to_input_directory>"
output_directory = "<path_to_output_directory>"

# Create an empty list to hold the names of files in the input directory
names = []

# Iterate through the files in the input directory and extract their names (without extensions)
for file in os.listdir(input_directory):
    names.append(file[:-5])

counter = 0

# Iterate through the files in the specified folder path
for filename in os.listdir(folder_path):
    # Check if the file is a .vtt file and if its name (without extension) is in the names list
    for file in names:
        # If the file name matches, parse the .vtt file and save the subtitles to a JSON file
        if filename.endswith(".vtt") and filename[:-4] == file:
            # Create the full path to the .vtt file and parse it
            counter += 1
            file_path = os.path.join(folder_path, filename)
            subtitles = parse_vtt_subtitles(file_path)
            # Create the output path for the JSON file and save the subtitles
            path = os.path.join(output_directory, filename[:-4]+".json")
            save_to_json(subtitles, path)
            print(counter , "     ", path)
```
